1. **From the given DataFrame find the relationship between experience and salary for both type of Company(Product Based and Service Based) using Line Plot.NOTE:salary(lakh/annum) Experience(years)** 

**Requirements:**
 xlabel="Experience in years",ylabel="Salary In lakh/annum",title="EXP vs SALARY".
 
 **NOTE:Use proper Styling and color for line plot**
 

```python
#Create a Dictionary of series
d1 = {
   'Experience':[2,3,4,5,6,7,8],
   'Salary':[10,12,14,18,20,26,33]
    }
d2 = {
   'Experience':[2,3,4,5,6,7,8],
   'Salary':[3.4,4,5,5.5,6,7,8]
    }
#Create a DataFrame
df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)
print (df1)
print(df2)
```

```python
#write the code in the cell
ex1=df1['Experience']
s1=df1['Salary']
ex2=df2['Experience']
s2=df2['Salary']
plt.plot(ex1,s1,color="blue", linestyle = "--", marker = "D",label='Product Based')
plt.plot(ex2,s2,color="red", linestyle = "-.", marker = "o",label='Service Based')
plt.title("EXP vs SALARY") # add the title
plt.xlabel("Experience in years") # add xlabel
plt.ylabel("Salary In lakh/annum")
plt.legend()
plt.show()
```

5. **Find the value of y for each value of x using the equation Y=5x+10 where x=[2,4,7,1,5,9,12,10].Represent the value of X and Y using the scatter plot**

```python
#write the code here
x=[2,4,7,1,5,9,12,10]
y= [(5*i)+10 for i in x]
plt.scatter(x,y,s=100,color="lime",edgecolors='black', linewidths=1)
plt.title("SCATTER PLOT OF EUATION Y=5X+10")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.xticks(ticks=x)
plt.show()
```

**6.  Draw the three different sub plots with the following equation. Plot x values on the x-axis and y-values on the y-axis. Give the neccessary labels in the plot. You can select x values of your choice. calculate y for atleast 50 values of x**

y1 = x^2 + 3x + 4 <br>
y2 = sigmoid(x) #please search the formula of sigmoid function<br>
y3 = logx <br>

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-100,100,50)

y1 =  x ** 2 + 3 * x + 4
y2 = 1/(1 + np.exp(-x))    #sigmod(x) =1/1+e^(-x)
y3 = np.log(x)

fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()
fig3, ax3 = plt.subplots()
 
ax1.plot(x,y1,color="mediumspringgreen",label="Y values on x")
ax2.plot(x,y2, linestyle = "--",color="lime",label="sigmod(x) values on x")
ax3.plot(x,y3, linestyle = "-.",color="m",label="log(x) values on x")
 
ax1.legend()
ax1.set_title("X VS Y")
ax1.set_xlabel("X")
ax1.set_ylabel("Y= X^2+3X+4")
 
ax2.legend()
ax2.set_xlabel("X")
ax2.set_ylabel("Y=Sigmod(X)")

ax3.legend()
ax3.set_xlabel("X")
ax3.set_ylabel("Y=log(x)")

plt.tight_layout()
plt.show()
```
